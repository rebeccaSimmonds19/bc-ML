```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

dataset = load_breast_cancer()
X = dataset.data

y = dataset.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=12345)
```

```python
y_pred = KMeans(n_clusters=2).fit_predict(X_train)

plt.figure(figsize=(12, 12))
scatter = plt.scatter(X_train[:,0], X_train[:,1], c=y_pred)
formatter = plt.FuncFormatter(lambda i, *args: dataset.target_names[int(i)])
classes = ['benign','malignant']
plt.colorbar(ticks=[0, 1], format=formatter)
plt.title("Benign vs malignant tumours", fontsize=20)

plt.show()
```
